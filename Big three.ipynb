```python
import datetime
import pandas_datareader
import quandl
from matplotlib.finance import candlestick_ohlc
from matplotlib.dates import DateFormatter, date2num, WeekdayLocator, DayLocator, MONDAY
```

```python
Etherium = quandl.get('GDAX/ETH_USD.1')
Bitcoin = quandl.get('BITFINEX/BTCUSD.1')
Litecoin = quandl.get('GDAX/LTC_USD.1')

Litecoin_volume = quandl.get('GDAX/LTC_USD.4')
Bitcoin_volume = quandl.get('BITFINEX/BTCUSD.7')
Etherium_volume = quandl.get('GDAX/ETH_USD.4')


```

```python

Etherium['MA50'] = Etherium['Open'].rolling(50).mean()
Etherium['MA200'] = Etherium['Open'].rolling(200).mean()
Etherium[['Open','MA50','MA200']].plot(label='Etherium',figsize=(16,8))



Etherium_volume.plot(title='Volume',figsize=(18,3))

;
```

```python
Bitcoin['MA50'] = Bitcoin['High'].rolling(50).mean()
Bitcoin['MA200'] = Bitcoin['High'].rolling(200).mean()
Bitcoin[['High','MA50','MA200']].plot(label='Bitcoin',figsize=(16,8))
Bitcoin_volume.plot(title='Volume',figsize=(18,3));
```

```python
Litecoin['MA50'] = Litecoin['Open'].rolling(50).mean()
Litecoin['MA200'] = Litecoin['Open'].rolling(200).mean()
Litecoin[['Open','MA50','MA200']].plot(label='Litecoin',figsize=(16,8))
Litecoin_volume.plot(title='Volume',figsize=(18,3));
```
